# Section 1: Creating a DataFrame from a List of Tuples

## 2. Creating a DataFrame from a List of Tuples

One of the simplest ways to create a Spark DataFrame is from a Python list of tuples.

Each tuple represents one row.

For example, this tuple:

```python
(1, "Alice", "Data Science", 95000)
```
could represent one employee with:
```
employee ID
name
department
salary
```
This is useful for learning because we can create small toy datasets directly inside the notebook.


````python
# Create a small list of tuples.
# Each tuple represents one row of employee data.

employee_data = [
   (1, "Alice", "Data Science", 95000),
   (2, "Bob", "Engineering", 88000),
   (3, "Carlos", "Engineering", 72000),
   (4, "Diana", "Operations", 68000),
   (5, "Evan", "Data Science", 105000)
]

# Create a Spark DataFrame without specifying column names.
# Spark will create default column names like _1, _2, _3, and _4.

employees_no_columns_df = spark.createDataFrame(employee_data)
````

````python
# show() prints the DataFrame in a text table.
# This is a Spark action, meaning it triggers execution.

employees_no_columns_df.show()
````

````python
# printSchema() shows the column names and data types.
# Since we did not provide column names, Spark uses default names.

employees_no_columns_df.printSchema()
````

# Section 2: Adding Column Names

## Important Observation

Spark created the DataFrame, but the column names are not meaningful.

Instead of names like:

- `employee_id`
- `name`
- `department`
- `salary`

Spark used default names:

- `_1`
- `_2`
- `_3`
- `_4`

This is why we usually provide column names when creating a DataFrame.

## 3. Adding Column Names

Now let's create the same DataFrame again, but this time we will provide meaningful column names.

This makes the DataFrame easier to understand and easier to query.

````python
# Define column names for the employee data.

employee_columns = ["employee_id", "name", "department", "salary"]

# Create a Spark DataFrame with column names.

employees_df = spark.createDataFrame(employee_data, employee_columns)
````

````python
# display() is Databricks' interactive table display function.
# It is very useful inside notebooks.

display(employees_df)
````

````python
# show() gives a simple text-based view of the DataFrame.

employees_df.show()
````

````python
# Print the schema.
# Spark inferred the data types from the Python values.

employees_df.printSchema()
````

````python
# The columns attribute returns a Python list of column names.

employees_df.columns
````

````python
# dtypes returns the column names and Spark data types.

employees_df.dtypes
````

# Section 3: Schema Basics

## 4. What Is a Schema?

A schema describes the structure of a DataFrame.

It tells Spark:

- The column names
- The data type of each column
- Whether each column can contain null values

For example, our employee DataFrame has a schema similar to this:

| Column | Data Type |
|---|---|
| employee_id | integer or long |
| name | string |
| department | string |
| salary | integer or long |

Schemas matter because Spark needs to understand the structure of the data before it can process it efficiently.


````python
# Let's inspect the schema object directly.

employees_df.schema
````

````python
# Print each field in the schema one at a time.
# This makes it easier to see the column name, data type, and nullable setting.

for field in employees_df.schema.fields:
   print("Column name:", field.name)
   print("Data type:", field.dataType)
   print("Nullable:", field.nullable)
   print("-" * 30)
````

# Section 4: Inferring Schema

## 5. Inferring Schema

When we created the employee DataFrame, Spark inferred the schema automatically.

That means Spark looked at the Python values and guessed the appropriate Spark data types.

For example:

| Python Value | Spark Type |
|---|---|
| `1` | LongType |
| `"Alice"` | StringType |
| `95000` | LongType |
| `95.5` | DoubleType |
| `True` | BooleanType |

Schema inference is convenient for small examples.

However, in production workflows, you often want to define the schema manually to avoid surprises.


````python
# Create a small dataset with mixed data types.
# Spark will infer the schema based on these Python values.

product_data = [
   (101, "Laptop", "Electronics", 999.99, True),
   (102, "Desk Chair", "Furniture", 149.50, True),
   (103, "Notebook", "Office Supplies", 3.99, False),
   (104, "Monitor", "Electronics", 249.99, True)
]

product_columns = ["product_id", "product_name", "category", "price", "in_stock"]

products_df = spark.createDataFrame(product_data, product_columns)

display(products_df)
````

````python
# Inspect the inferred schema.

products_df.printSchema()
````

````python
# View the column names and data types as a list.

products_df.dtypes
````

# Section 5: Defining Schema Manually

## 6. Defining a Schema Manually

Sometimes you do not want Spark to guess the data types.

Instead, you can define the schema yourself.

This gives you more control.

A manually defined schema is useful when:

- Reading messy data
- Working with dates and timestamps
- Making sure IDs are strings instead of numbers
- Controlling decimal, integer, or double types
- Building reliable data pipelines

To manually define a schema, we use classes from `pyspark.sql.types`.


````python
# Import Spark SQL data types.
# These allow us to define a schema manually.

from pyspark.sql.types import (
   StructType,
   StructField,
   IntegerType,
   LongType,
   StringType,
   DoubleType,
   BooleanType
)
````

````python
# Define a manual schema for the product data.
# StructType represents the full schema.
# StructField represents one column.

product_schema = StructType([
   StructField("product_id", IntegerType(), nullable=False),
   StructField("product_name", StringType(), nullable=False),
   StructField("category", StringType(), nullable=True),
   StructField("price", DoubleType(), nullable=True),
   StructField("in_stock", BooleanType(), nullable=True)
])
````

````python
# Create a DataFrame using the manually defined schema.

products_manual_schema_df = spark.createDataFrame(
   product_data,
   schema=product_schema
)

display(products_manual_schema_df)
````

````python
# Inspect the manually defined schema.
# Notice that product_id is now IntegerType instead of whatever Spark inferred.

products_manual_schema_df.printSchema()
````

## Inferred Schema vs Manual Schema

Schema inference is fast and convenient for quick exploration.

Manual schemas are better when correctness matters.

| Approach | Best For |
|---|---|
| Inferred schema | Quick experiments and small toy examples |
| Manual schema | Reliable pipelines and controlled data types |

For learning notebooks, inferred schemas are usually fine.

For real data engineering work, manual schemas are often safer.

# Section 6: Working with Null Values

## 7. Creating a DataFrame with Null Values

Real datasets often contain missing values.

In Python, missing values are usually represented as `None`.

When Spark sees `None`, it treats it as a null value.

Let's create a small customer dataset with some missing values.

````python
# Create customer data with some missing values.
# None becomes null in Spark.

customer_data = [
   (1, "Alice", "Gainesville", "FL"),
   (2, "Bob", None, "FL"),
   (3, "Carlos", "Atlanta", "GA"),
   (4, "Diana", "Orlando", None),
   (5, None, "Tampa", "FL")
]

customer_columns = ["customer_id", "customer_name", "city", "state"]

customers_df = spark.createDataFrame(customer_data, customer_columns)

display(customers_df)
````

````python
# Inspect the schema.
# Spark marks these fields as nullable because null values may exist.

customers_df.printSchema()
````

````python
# Count the number of rows in the DataFrame.
# count() is an action.

customers_df.count()
````

# Section 7: Common DataFrame Inspection Methods

## 8. Common DataFrame Inspection Methods

When you create a DataFrame, you should inspect it before doing analysis.

Useful methods include:

| Method | Purpose |
|---|---|
| `display(df)` | Databricks interactive table view |
| `df.show()` | Text table output |
| `df.printSchema()` | View column names and data types |
| `df.columns` | Get column names |
| `df.dtypes` | Get column names and data types |
| `df.count()` | Count rows |
| `df.limit(n)` | Preview first n rows |

````python
# Show the first few rows as a text table.

customers_df.show()
````

````python
# Show only the first 3 rows.
# limit() is a transformation.
# show() is the action that triggers execution.

customers_df.limit(3).show()
````

````python
# Get the column names.

customers_df.columns
````

````python
# Get the data types.

customers_df.dtypes
````

````python
# Count rows.

row_count = customers_df.count()

print("Number of customer rows:", row_count)
````

# Section 8: Mini-Project — Retail Transaction DataFrame

Now we will create a small retail transaction DataFrame.

This dataset will include:

| Column | Meaning |
|---|---|
| transaction_id | Unique transaction identifier |
| store_id | Store where the transaction occurred |
| customer_id | Customer identifier |
| amount | Transaction amount |
| category | Product category |
| timestamp | Time of transaction |

This type of dataset is useful because it resembles the kind of structured data you might analyze in retail analytics, fraud detection, customer analytics, or loss prevention work.

````python
# Import TimestampType for timestamp columns.

from pyspark.sql.types import TimestampType
from datetime import datetime
````

````python
# Create a small list of retail transactions.
# Each tuple represents one transaction.

retail_transaction_data = [
   (1001, "Store_001", "Customer_001", 249.99, "Electronics", datetime(2026, 5, 1, 9, 15, 0)),
   (1002, "Store_001", "Customer_002", 34.50, "Grocery", datetime(2026, 5, 1, 10, 5, 0)),
   (1003, "Store_002", "Customer_003", 89.99, "Clothing", datetime(2026, 5, 1, 11, 30, 0)),
   (1004, "Store_002", "Customer_001", 599.99, "Electronics", datetime(2026, 5, 2, 14, 45, 0)),
   (1005, "Store_003", "Customer_004", 22.25, "Grocery", datetime(2026, 5, 2, 16, 20, 0)),
   (1006, "Store_003", "Customer_005", 129.99, "Home Goods", datetime(2026, 5, 3, 13, 10, 0)),
   (1007, "Store_001", "Customer_006", 14.99, "Grocery", datetime(2026, 5, 3, 17, 55, 0)),
   (1008, "Store_002", "Customer_007", 799.99, "Electronics", datetime(2026, 5, 4, 12, 0, 0))
]
````

````python
# Define a manual schema for the retail transaction data.
# This gives us full control over the column names and data types.

retail_transaction_schema = StructType([
   StructField("transaction_id", IntegerType(), nullable=False),
   StructField("store_id", StringType(), nullable=False),
   StructField("customer_id", StringType(), nullable=False),
   StructField("amount", DoubleType(), nullable=False),
   StructField("category", StringType(), nullable=True),
   StructField("timestamp", TimestampType(), nullable=False)
])
````

````python
# Create the retail transaction Spark DataFrame.

transactions_df = spark.createDataFrame(
   retail_transaction_data,
   schema=retail_transaction_schema
)

display(transactions_df)
````

````python
# Inspect the schema.

transactions_df.printSchema()
````

````python
# View column names.

transactions_df.columns
````

````python
# View data types.

transactions_df.dtypes
````

````python
# Count the number of transactions.
# This is an action.

transaction_count = transactions_df.count()

print("Number of transactions:", transaction_count)
````

# Section 9: Quick Exploration of the Retail DataFrame

## 9. Quick Exploration

Even though this notebook is mainly about creating DataFrames, let's do a few simple checks.

We will answer:

1. How many transactions are in the dataset?
2. What categories are present?
3. What are the highest-value transactions (and limit to top 3)?

# Section 10: Create a DataFrame from Dictionaries

## 10. Creating a DataFrame from a List of Dictionaries

You can also create a Spark DataFrame from a list of Python dictionaries.

Each dictionary represents one row.

The dictionary keys become column names.

This can be convenient, but for larger or more reliable workflows, tuples plus a schema are often cleaner.


````python
# Create a list of dictionaries.
# Each dictionary represents one row of store data.

store_data = [
   {"store_id": "Store_001", "city": "Gainesville", "state": "FL", "store_type": "Urban"},
   {"store_id": "Store_002", "city": "Orlando", "state": "FL", "store_type": "Suburban"},
   {"store_id": "Store_003", "city": "Atlanta", "state": "GA", "store_type": "Urban"}
]

stores_df = spark.createDataFrame(store_data)

display(stores_df)
````

````python
# Inspect the inferred schema.

stores_df.printSchema()
````

# Section 11: Common Mistakes

# Common Mistakes When Creating DataFrames

## Mistake 1: Forgetting Column Names

If you create a DataFrame without column names, Spark will use default names like `_1`, `_2`, and `_3`.

That makes your code harder to read.

## Mistake 2: Relying Too Much on Schema Inference

Schema inference is convenient, but Spark may infer a type you did not expect.

Manual schemas are safer when correctness matters.

## Mistake 3: Mixing Data Types in the Same Column

Try to keep each column consistent.

For example, do not mix numbers and strings in the same column unless you truly want the column to be treated as a string.

## Mistake 4: Forgetting That Spark DataFrames Are Distributed

A Spark DataFrame may look like a local object, but it represents distributed data and a logical execution plan.

Do not assume it behaves exactly like a pandas DataFrame.
